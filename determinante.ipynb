## Definição de determinante:
Fixada uma linha $i$ de uma matriz quadrada $A_{nxn} = (a_{ij}) \in M_n(\mathbb{R})$, definimos o determinante de $A$ pela expressão:

$$ \sum_{j=1}^n a_{ij} \cdot (-1)^{i+j} \cdot det\text{A}_{-i,-j}$$

Onde $\text{A}_{-i,-j}$ é a matriz resultante da remoção da linha $i$ e coluna $j$

```python
import numpy as np
import sys
```

```python
def calcular_determinante(matriz):
    detM = 0
    (ordem, _) = matriz.shape
    
    if ordem == 2:
        return determinante_2(matriz)
    for i in range(0, ordem):
        sub_matriz = np.delete(np.delete(matriz, 0, 0), i, 1)
        print(sub_matriz)
        determinante_inferior = calcular_determinante(sub_matriz)
        detM += matriz[0][i] * (-1)**(i) * determinante_inferior
        print(matriz[0][i] * (-1)**(i))
    return detM

def determinante_2(matriz):
    dp = 1
    ds = 1 
    for i in matriz.diagonal():
        dp *= i
    for i in matriz[::-1].diagonal():
        ds *= i
    return dp - ds
```
